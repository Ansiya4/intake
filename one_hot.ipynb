```python
import pandas as pd

data = {
    'Age': [25, 35, 45, 30],
    'Gender': ['Male', 'Female', 'Male', 'Female'],
    'Income': [50000, 75000, 90000, 60000],
    'Education Level': ['High School', "Bachelor's", "Master's", "Bachelor's"],
    'Purchased': ['No', 'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data)
print(df)
```

```python
cate=['Gender','Education Level','Purchased']
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one= OneHotEncoder()
```

```python
encoded_col=one.fit_transform(df[cate])
```

```python
one_df=pd.DataFrame(encoded_col.toarray(),columns=one.get_feature_names_out(cate))
```

```python
one_df
```

```python
df_encoded=pd.concat([df.drop(columns=cate),one_df],axis=1)
```

```python
df_encoded
```
